```python
geolocation.geolocation_city.value_counts()
```

```python
#Se eliminan caracteres especiales.
replacements = (
        ("á", "a"),
        ("ã", "a"),
        ("à","a"),
        ("å","a"),
        ("â","a"),
        ("ä","a"),
        ("é", "e"),
        ("ê","e"),
        ("ë","e"),
        ("è","e"),
        ("í", "i"),
        ("ï","i"),
        ("î","i"),
        ("ì","i"),
        ("ó", "o"),
        ("ô","o"),
        ("ö","o"),
        ("ò","o"),
        ("ú", "u"),
        ("û","u"),
        ("ù","u"),
        ("ç", "c"),
        ("£", ""),
        ("-"," "),
        ("saopaulo", "sao paulo")
    )
def replace_special_characters(text):
    for letter in replacements:
        text = text.replace(letter[0],letter[1])
    if text == 'sp':
        text = 'sao paulo'
    if ',' in text:
        if 'rio de janeiro' in text:
            text = 'rio de janeiro'
        if 'campo alegre de lourdes' in text:
            text = 'campo alegre de lourdes'
    return text


geolocation.geolocation_city = geolocation.geolocation_city.map(lambda x : replace_special_characters(x))
```

```python
# calculamos el promedio de la diferencia entre order_purchase_timestamp y order_approved_at para completar nulos depues
orders_purchase_approved = orders[['order_approved_at','order_purchase_timestamp']].loc[orders['order_approved_at'].isnull() == False]
orders_purchase_approved['order_approved_at'] =  pd.to_datetime(orders['order_approved_at'])
orders_purchase_approved_prom = (orders_purchase_approved['order_approved_at'] - orders_purchase_approved['order_purchase_timestamp'])
orders_purchase_approved_prom = orders_purchase_approved_prom.mean()
```

```python
orders.order_approved_at.fillna(value = (orders['order_purchase_timestamp'] + orders_purchase_approved_prom),inplace=True)
orders['order_approved_at'] = pd.to_datetime(orders.order_approved_at)
```

### columna order_delivered_customer_date y order_delivered_carrier_date

```python
filt = orders[(orders['order_status'] == 'delivered') & ((orders['order_delivered_customer_date'].isnull()) | (orders['order_delivered_carrier_date'].isnull()))].index
orders.drop(filt,inplace = True)
# las ordenes que figuran como canceladas pero fueron entregadas, procedemos a eliminarlas porque es un caso muy raro
filt = orders[(orders['order_status'] != 'delivered') & (orders['order_delivered_customer_date'].isnull() == False)].index
orders.drop(filt,inplace = True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

### RFM analisis:
voy a buscar a los mejores customers teniendo en cuenta 3 variables: Recency(lo reciente que haya sido su ultimo movimiento), frequency (la frecuencia con la que realiza comvimientos) y monetary (ingresos que generó)

#### RECENCY

```python
#primero establezco el dia actual como la fecha de la ultima order que figure en la tabla orders y le sumare un dia
current_date = orders['order_purchase_timestamp'].max() + dt.timedelta(days = 1)

orders_X_customers = pd.merge(orders,customers, on = 'customer_id')
# elimino los que tengan su ultima compra hace mas de un año 
orders_X_customers = orders_X_customers[(current_date - orders_X_customers['order_purchase_timestamp']) < np.timedelta64(365, 'D')]
recency = orders_X_customers.groupby('customer_unique_id',as_index=False)['order_purchase_timestamp'].max()
recency['recency'] = recency['order_purchase_timestamp'].apply(lambda x : (current_date-x).days)
recency.drop(columns='order_purchase_timestamp',inplace = True)
recency
```

#### FREQUENCY

```python
frequency = orders_X_customers[['customer_unique_id']]
frequency =frequency.groupby('customer_unique_id', as_index=False).size()
frequency.rename(columns={'size': 'frequency'},inplace=True)
frequency.sort_values('frequency')
```

#### MONETARY

```python
monetary = pd.merge(orders_X_customers,order_items, on = 'order_id')
monetary = monetary.groupby('customer_unique_id', as_index=False)['price'].sum()
monetary.rename(columns={'price': 'monetary'},inplace=True)
monetary
```

#### distribucion de los valores

```python
best_customers = pd.merge(frequency,recency, on='customer_unique_id')
best_customers = pd.merge(best_customers,monetary, on='customer_unique_id')
best_customers
```

### RFM score

priemro voy a segmentar los datos con valores entre 1 y 5

```python
best_customers["recency_score"] = pd.qcut(best_customers['recency'], 5, labels=[5, 4, 3, 2, 1])
best_customers['recency_score'] = best_customers['recency_score'].astype('int64')
best_customers["monetary_score"] = pd.qcut(best_customers['monetary'], 5, labels=[1, 2, 3, 4, 5])
best_customers['monetary_score'] = best_customers['monetary_score'].astype('int64')
```

```python
best_customers['frequency'].value_counts(normalize=True)*100
```

frequency tiene casi todos sus datos en 1 asi que los dividire manualmente: si la frequency es mayor o igual a 5 dejo 5 y si es menor dejo el mismo drequency

```python
def parse_values(x):
    if x>5:
        return 5
    else:
        return x
```

```python
best_customers['frequency_score'] = best_customers['frequency'].apply(lambda row : parse_values(row))
```

ahora para calcualr el score voy a formar un numero concatenando los resultados en el orden:
(recency)-> (frequency)-> (monetary)

 + best_customers['frequency_score']*10 + best_customers['monetary_score']
best_customers

```python
best_customers.info()
```

```python
best_customers['RFM_score'] = best_customers['recency_score']*100 + best_customers['frequency_score']*10 + best_customers['monetary_score']
best_customers.sort_values('RFM_score',ascending=False)
```
